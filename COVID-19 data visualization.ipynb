```
import requests
import pandas as pd
```

```
x = requests.get('https://api.covid19api.com/summary').json()
print(x)
```

```
df = pd.DataFrame.from_dict(x['Countries'])
df
```

```
country_wise_grouped = df.groupby("Country").mean()

country_wise_grouped.sort_values(by=['TotalConfirmed'], ascending=False)[:10].plot(kind="bar")
```

```
country_wise_grouped.sort_values(by=['NewConfirmed'], ascending=False)[:10].plot(kind="bar")
```

```
country_wise_grouped.sort_values(by=['TotalRecovered'], ascending=False)[:10].plot(kind="bar", y=['TotalRecovered','NewConfirmed'])
```
